- Crie uma lista


```python
lista_cem = list(range(101))
```

- Crie um for loop para imprimir na tela todos os elementos da lista


```python
for item in lista_cem:
    print(item)
```

- Crie um for loop usando o recurso "continue"


```python
for item in lista_cem:
    if item % 2 == 0:
        continue
    print(item)
```

- Crie um for loop usando o recurso "break"


```python
for item in lista_cem:
    if item == 98:
        break
    print(item)
```

- Crie um for loop usando o "else"


```python
for item in lista_cem:
    if item % 2 == 0:
        continue
    print(item)
else:
    print("Lista de impares impressa!")
```
